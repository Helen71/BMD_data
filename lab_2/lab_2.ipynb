```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('title')
```

2.Побудова графіків

```python
plt.subplot(1,2,1)
plt.plot (x,y,'r')

plt.subplot(1,2,2)
plt.plot (y,x,'b')
```

2.2 Об`єктно-орієнтований підхід

```python
fig=plt.figure()

axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel('x label')
axes.set_ylabel('y label')
axes.set_title('title')
```

```python
fig=plt.figure()

axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.1,0.5,0.4,0.3])


axes1.plot(x,y,'r')
axes1.set_xlabel('x label')
axes1.set_ylabel('y label')
axes1.set_title('title')

axes2.plot(y,x,'b')
axes2.set_xlabel('x label')
axes2.set_ylabel('y label')
axes2.set_title('title')
```

```python
fig, axes=plt.subplots()

axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
```

```python
fig, axes=plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig, axes=plt.subplots(nrows=4, ncols=4) 
plt.tight_layout()
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])

ax.plot(x,x**2, label="x**2")
ax.plot(x,x**3, label="x**3")
ax.legend()
```

```python
fig=plt.figure(figsize=(8,4), dpi=100)
```

```python
fig,ax=plt.subplots()
ax.plot(x,x**2, color="b")
ax.plot(x,x**3, color="g")
```

```python
fig,ax=plt.subplots()
ax.plot(x,x**3, lw=2, ls='--')
```

```python
fig,ax=plt.subplots()
ax.plot(x,x**2, marker='o', markersize=10, markerfacecolor='y')
```
