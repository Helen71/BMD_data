```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Завантаження вбудованих наборів даних:

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

Візуалізація даних.
Графіки для аналізу кількісних даних:

1.displot. Гістограма та графік щільності розподілу

```python
sns.distplot(tips['total_bill'])
```

2.jointplot. Створює діаграму розсіювання, за допомогою якої можна зіставити розподіл двох змінних та їх взаємозв’язок.

```python
sns.jointplot( x='total_bill', y='tip', data=tips,kind="kde", color="#4ca8ed", height = 7, xlim = (0,50), ylim = (0,7))
```


3.pairplot – будує попарні взаємозв’язки між змінними в повному наборі даних.

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, palette='sex')
```

```python
sns.pairplot(tips, hue='sex')
```

Графіки для аналізу категоріальних даних:

1.barplot – дозволяють отримувати сукупні дані по кількісним даним в залежності від категоріальної змінної.

```python
sns.barplot(x='sex',y='total_bill', data=tips)
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.median)
```

```python

sns.barplot(x = "day", y = "tip", data = tips,palette="rocket", estimator=np.median)
```

2.boxplots - використовується для візуалізації розподілу категоріальних даних за певною кількісною змінною.

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
sns.boxplot(x = "day", y = "total_bill",hue="smoker", data = tips, palette="pink")
```

```python
sns.boxplot(data = tips, palette="pink", orient='h')
```

3.stripplot – представляє собою діаграму розсіювання, де одна за змінних категоріальна.

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

Графіки для аналізу взаємозв’язку даних:

Heatmap – графік взаємозв’язку (кореляції) між даними у вигляді матриці.

```python
sns.heatmap(tips.corr())
```

Графіки для відображення даних у вигляді сітки:

```python
g = sns.FacetGrid(tips, col = "time", row = "smoker", hue = "day")
g = g.map(plt.hist, "total_bill")
g.add_legend(title="legend")
```
